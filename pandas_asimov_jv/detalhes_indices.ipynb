```python
import pandas as pd
import numpy as np
from numpy.random import randn
```

```python
#posso saber os indices do meu df com o df.index e para as colunas df.columns

df = pd.DataFrame(randn(5,4), index = ['A','B','C','D','E'], columns = ['price','date', 'brand','profit'])
print(df.index)
print(df.columns)
```

```python
#se eu usar o metodo df.reset_index(), ele ira criar indices de 0 a n linhas, 
# o indice original vira uma nova coluna no df
#OBS: note que novamente usa-se o inplace  = True para poder alterar mesmo no df original


df.reset_index(inplace=True)
df
```

```python
#para voltar o indice original eu posso usar o: 
#  df.set_index("nome da coluna para virar o indice", inplace = True)

df.set_index("index", inplace = True)
df
```

```python
#criando e setando um novo indice
#obs: NESSE metodo, o novo indice Ã© deletado e o indice setado vira o padrao
estados = 'RJ SP ES MG SC'.split()

df['novo_indice']= estados
df.set_index('novo_indice', inplace = True)
df
```

```python
df = pd.DataFrame(randn(5,4), index = ['A','B','C','D','E'], columns = ['price','date', 'brand','profit'])
```

```python
# para manter o indice atual, temos que dar um df.reset_index(), 
# para ele guardar a info do indice em uma coluna

estados = 'RJ SP ES MG SC'.split()

df['novo_indice']= estados
df.reset_index().set_index('novo_indice')
df
```
